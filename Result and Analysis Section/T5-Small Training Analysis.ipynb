```python
import pandas as pd
import matplotlib.pyplot as plt

with open("t5_small on all dataset training.txt", 'r', encoding='utf-8') as f:
    lines = f.readlines()
```

```python
train_loss = []
for line in lines:
    if 'Iteration' in line:
        train_loss.append(float(line.split(' ')[-1].strip('\n')))
plt.figure(figsize = (10, 4))
plt.plot(train_loss)
plt.show()
```
